```python
df = pd.read_csv('../mama.csv', sep='\t', dtype=str)
# 칼럼 분리
df.columns = ['createdAt,data']  # 현재 하나의 칼럼명
df = df['createdAt,data'].str.split(',', n=1, expand=True)
df.columns = ['createdAt', 'data']
df
```

```python
def parse_json(json_str, created_at=None):
    try:
        # 1. JSON 형식 수정
        json_str = json_str.replace('"{"result":', '{"result":')
        
        # 2. 마지막 부분 수정 (마지막 따옴표 제거)
        if json_str.endswith('}]}"'):
            json_str = json_str[:-1]  # 마지막 따옴표만 제거
            
        # 3. JSON 파싱
        data = json.loads(json_str)
        result_df = pd.DataFrame(data['result'])
        # 4. createdAt 컬럼 추가
        if created_at is not None:
            result_df['createdAt'] = created_at
        
        return result_df
    except Exception as e:
        print("파싱 에러:", str(e))
        return None

# 첫 번째 행 파싱 시도
print("첫 번째 행 파싱 결과:")
first_row_data = parse_json(df['data'].iloc[0])

if first_row_data is not None:
    print("\n데이터 형태:", first_row_data.shape)
    print("\n처음 몇 개의 행:")
    print(first_row_data[['countryNameKr', 'currencyCode', 'basePrice']].head())
```

```python
# 전체 데이터 처리
print(f"데이터 수: {len(df)}")

# 2. 모든 행의 JSON 데이터 파싱
parsed_data = []
for idx, row in df.iterrows():
    result = parse_json(row['data'], row['createdAt'])
    if result is not None:
        # createdAt 칼럼 한국 시간대로 변경
        result['createdAt'] = pd.to_datetime(result['createdAt'], format='%Y-%m-%d %H:%M:%S') + pd.Timedelta(hours=9)
        parsed_data.append(result)

# 3. 모든 파싱된 데이터를 하나의 데이터프레임으로 합치기
final_df = pd.concat(parsed_data, ignore_index=True)

# 4. 결과 확인
print("\n최종 데이터프레임 형태:", final_df.shape)
print("\n주요 통화 환율 추이:")
main_currencies = ['USD', 'JPY', 'EUR', 'CNY']
print(final_df[final_df['currencyCode'].isin(main_currencies)][['countryNameKr', 'currencyCode', 'basePrice', 'createdAt']].head(12))
```

1. isBuyOrder = 1 일 때, price -1 이 등장하는가?
2. isBuyOrder = 0 일 때, price +1 이 등장하는가?

```python
# 매수 주문 필터링
buy = trade[trade['isBuyOrder'] == 1]

# 매수 주문에 대해 등장 여부 확인
matches = 0

for _, order in buy.iterrows():
    # 동일한 통화 / price +1
    match_count = filtered_df[
        (filtered_df['currencyCode'] == order['currencyCode']) &
        (filtered_df['basePrice'] == order['price'] + 1)
    ].shape[0]
    matches +=match_count

print(f'매수 주문 등장 횟수 : {matches}')
```

```python
# 더 자세한 분석을 위해 통화별로 확인
print("\n통화별 매칭 횟수:")
for currency in filtered_df['currencyCode'].unique():
    currency_buy_orders = buy[buy['currencyCode'] == currency]
    currency_matches = 0
    
    for _, order in currency_buy_orders.iterrows():
        match_count = filtered_df[
            (filtered_df['currencyCode'] == currency) & 
            (filtered_df['basePrice'] == order['price'] + 1)
        ].shape[0]
        currency_matches += match_count
    
    print(f"{currency}: {currency_matches}회")
```

```python
# 결과를 저장할 리스트 초기화
results = []

# trade의 각 행을 순회
for idx, trade_row in trade.iterrows():
    # KRW인 경우 currencyCode0 사용
    currency = trade_row['currencyCode0'] if trade_row['currencyCode'] == 'KRW' else trade_row['currencyCode']
    trade_row['executedAt'] = pd.to_datetime(trade_row['executedAt'], format='%Y-%m-%d %H:%M:%S') + pd.Timedelta(hours=9)
    # 매수 /매도에 다라 target_price 계산
    if trade_row['isBuyOrder'] == 1: # 매수
        target_price = trade_row['price'] - 1
    else: # 매도
        target_price = trade_row['price'] + 1
        # target price가 filtered_df에 존재하는지 확인
        matches = filtered_df[
            (filtered_df['currencyCode'] == currency) & 
            (filtered_df['basePrice'] == target_price)
        ].shape[0]

    
    results.append({
        'currency': currency,
        'order_type' : '매수' if trade_row['isBuyOrder'] == 1 else '매도',
        'original_price': trade_row['price'],
        'target_price': target_price,
        'found': matches > 0,
        'match_count': matches,
        'createdAt' : trade_row['executedAt']
    })

# 결과를 데이터프레임으로 변환
results_df = pd.DataFrame(results)

# 결과 요약
print("전체 거래 수:", len(results))
print("목표가가 등장한 거래 수:", results_df['found'].sum())

print("\n주문 유형별 매칭 결과:")
print(results_df.groupby('order_type')[['found']].sum())

print("\n통화별 매칭 결과:")
print(results_df.groupby('currency')[['found']].sum())

print("\n통화 및 주문 유형별 매칭 결과:")
print(results_df.groupby(['currency', 'order_type'])[['found']].sum())

# 전체 결과 확인
print("\n처음 10개 거래 상세 결과:")
results_df.head(10)
```

```python
start_date = '2024-08-01'
end_date = '2025-02-28'
# ✅ 데이터 다운로드 함수
def get_yfinance_data(ticker, start_date, end_date):
    data = yf.download(ticker, start=start_date, end=end_date, interval="1h")
    if isinstance(data, pd.DataFrame):
        return data[['Open', 'High', 'Low', 'Close']] if 'Open' in data.columns else data[['Close']]
    else:
        raise ValueError(f"Data for {ticker} is not a DataFrame. Got {type(data)}.")

# ✅ 1. USD/KRW 환율 데이터 가져오기
usd_krw = get_yfinance_data("USDKRW=X", start_date, end_date)

# ✅ 2. 금 가격 (Gold Futures) 가져오기
jpy_krw = get_yfinance_data("JPYKRW=X", start_date, end_date)

# ✅ 3. 원유 가격 (Crude Oil Futures) 가져오기
cad_krw = get_yfinance_data("CADKRW=X", start_date, end_date)


# ✅ 컬럼명 변경 (구분하기 쉽게)
usd_krw.columns = ['usd_krw_open', 'usd_krw_high', 'usd_krw_low', 'usd_krw_close']
jpy_krw.columns = ['jpy_krw_open', 'jpy_krw_high', 'jpy_krw_low', 'jpy_krw_close']
cad_krw.columns = ['cad_krw_open', 'cad_krw_high', 'cad_krw_low', 'cad_krw_close']

# ✅ 날짜 인덱스를 기준으로 병합 (merge 사용)
df = usd_krw.merge(jpy_krw, how='outer', left_index=True, right_index=True)
df = df.merge(cad_krw, how='outer', left_index=True, right_index=True)
df.reset_index(inplace=True)
df
```

```python
# ✅ 3. 데이터를 melt하여 통화별 데이터를 통합
melted_df = pd.melt(
    df,
    id_vars=['Datetime'],
    value_vars=[
        'usd_krw_open', 'usd_krw_high', 'usd_krw_low', 'usd_krw_close',
        'jpy_krw_open', 'jpy_krw_high', 'jpy_krw_low', 'jpy_krw_close',
        'cad_krw_open', 'cad_krw_high', 'cad_krw_low', 'cad_krw_close'
    ],
    var_name='metric',
    value_name='value'
)

# ✅ 4. 'metric'에서 통화 코드와 가격 종류(open, high, low, close) 분리
melted_df['currencyCode'] = melted_df['metric'].apply(lambda x: x.split('_')[0].upper())
melted_df['priceType'] = melted_df['metric'].apply(lambda x: x.split('_')[-1].lower())

# ✅ 5. 'metric' 열 삭제
melted_df.drop(columns=['metric'], inplace=True)

# ✅ 6. 피벗(pivot)하여 하나의 행에 고가, 저가, 오픈, 종가 정보가 들어가도록 변환
final_df = melted_df.pivot_table(
    index=['Datetime', 'currencyCode'],
    columns='priceType',
    values='value'
).reset_index()

# ✅ 컬럼 정리
final_df.columns.name = None  # 컬럼명 정리
final_df = final_df[['Datetime', 'currencyCode', 'high', 'low', 'open', 'close']]

# ✅ 결과 출력
print(final_df.head(10))
```

```python
def load_yahoo_data(currencies, start_date, end_date):
    """
    야후 파이낸스에서 통화 데이터를 다운로드합니다.
    """
    data_frames = []
    for currency in currencies:
        # 통화 코드 형식: "USDKRW=X", "JPYKRW=X" 등의 형태로 설정
        ticker = f'{currency}KRW=X'
        df = yf.download(ticker, start=start_date, end=end_date)
        df['currency'] = currency
        df.reset_index(inplace=True)
        df.rename(columns={
            'Date': 'createdAt',
            'High': 'highPrice',
            'Low': 'lowPrice',
            'Close': 'closePrice'
        }, inplace=True)
        data_frames.append(df)
    return pd.concat(data_frames, ignore_index=True)
```

```python
import yfinance as yf
import pandas as pd

def get_currency_data(currencies, start_date, end_date):
    data_frames = []

    for currency in currencies:
        ticker = f"{currency}KRW=X"
        data = yf.download(ticker, start=start_date, end=end_date)

        if isinstance(data, pd.DataFrame):
            data = data[['Open', 'High', 'Low', 'Close']] if 'Open' in data.columns else data[['Close']]
            data.columns = [f"{currency.lower()}_krw_open", f"{currency.lower()}_krw_high",
                            f"{currency.lower()}_krw_low", f"{currency.lower()}_krw_close"]
            data.reset_index(inplace=True)  # Datetime을 인덱스에서 열로 변환
            data_frames.append(data)
        else:
            raise ValueError(f"Data for {ticker} is not a DataFrame. Got {type(data)}.")

    # 병합
    merged_df = pd.concat(data_frames, axis=1)

    # ✅ 'Datetime' 열이 중복될 수 있으므로 병합 후 하나로 통일
    merged_df = merged_df.loc[:, ~merged_df.columns.duplicated()]
    merged_df['Date'] = pd.to_datetime(merged_df['Date']).dt.strftime('%Y-%m-%d')
    # 데이터 melt
    melted_df = pd.melt(
        merged_df,
        id_vars=['Date'],
        value_vars=[col for col in merged_df.columns if col != 'Date'],
        var_name='metric',
        value_name='value'
    )

    # 통화 코드 및 가격 타입 추출
    melted_df['currencyCode'] = melted_df['metric'].apply(lambda x: x.split('_')[0].upper())
    melted_df['priceType'] = melted_df['metric'].apply(lambda x: x.split('_')[-1].lower())

    # ✅ JPY 데이터만 100을 곱함
    melted_df.loc[melted_df['currencyCode'] == 'JPY', 'value'] *= 100

    # ✅ 모든 값 소수점 둘째 자리까지 표시
    melted_df['value'] = melted_df['value'].round(2)

    # 피벗하여 최종 데이터 프레임 생성
    final_df = melted_df.pivot_table(
        index=['Date', 'currencyCode'],
        columns='priceType',
        values='value'
    ).reset_index()

    # 컬럼 정리
    final_df.columns.name = None
    final_df = final_df[['Date', 'currencyCode', 'high', 'low', 'open', 'close']]

    return final_df

# ✅ 함수 사용 예제
start_date = '2024-08-01'
end_date = '2025-02-20'
currencies = ['USD', 'JPY', 'CAD']

currency_data = get_currency_data(currencies, start_date, end_date)
print(currency_data.tail(10))
```

```python
# 지정한 값
threshold = 1  # 예시로 1을 지정

# 결과 카운트
match_count = 0

# trade_df의 각 거래에 대해 처리
for index, trade in trade_df.iterrows():
    # 해당 거래에 대해 날짜와 통화 코드에 맞는 final_df의 행 찾기
    matching_data = final_df[(final_df['Date'] == trade['executedAt'].normalize()) & (final_df['currencyCode'] == trade['currencyCode'])]
    
    # 매칭되는 데이터가 있을 때
    if not matching_data.empty:
        high = matching_data['high'].values[0]
        low = matching_data['low'].values[0]
        
        # 매수 거래인지 매도 거래인지 확인
        if trade['isBuyOrder'] == 1:  # 매수
            if (trade['price'] - threshold >= low) and (trade['price'] - threshold <= high):
                match_count += 1
        else:  # 매도
            if (trade['price'] + threshold >= low) and (trade['price'] + threshold <= high):
                match_count += 1

print(f"조건을 만족하는 거래의 수: {match_count}")
```
