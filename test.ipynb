```python
# JSON 문자열 수정 및 파싱
def parse_json(json_str):
    try:
        # 1. JSON 형식 수정
        json_str = json_str.replace('{result":', '{"result":')
        
        # 2. 마지막 따옴표 처리
        if json_str.endswith('"}'):
            json_str = json_str[:-2] + '}'
            
        # 3. JSON 파싱
        data = json.loads(json_str)
        return pd.DataFrame(data['result'])
    except Exception as e:
        print("파싱 에러:", str(e))
        # 에러 발생 위치의 문자열 출력
        if hasattr(e, 'pos'):
            pos = e.pos
            print("문제 발생 부분:", json_str[max(0, pos-50):min(len(json_str), pos+50)])
        return None

# 첫 번째 행 파싱 시도
print("첫 번째 행 파싱 결과:")
first_row_data = parse_json(df['data'].iloc[0])

if first_row_data is not None:
    print("\n데이터 형태:", first_row_data.shape)
    print("\n처음 몇 개의 행:")
    print(first_row_data.head())
```

```python
df = pd.read_csv('../mama.csv', sep='\t', dtype=str)
# 칼럼 분리
df.columns = ['createdAt,data']  # 현재 하나의 칼럼명
df = df['createdAt,data'].str.split(',', n=1, expand=True)
df.columns = ['createdAt', 'data']
df
```

```python
def parse_json(json_str, created_at=None):
    try:
        # 1. JSON 형식 수정
        json_str = json_str.replace('"{"result":', '{"result":')
        
        # 2. 마지막 부분 수정 (마지막 따옴표 제거)
        if json_str.endswith('}]}"'):
            json_str = json_str[:-1]  # 마지막 따옴표만 제거
            
        # 3. JSON 파싱
        data = json.loads(json_str)
        result_df = pd.DataFrame(data['result'])
        # 4. createdAt 컬럼 추가
        if created_at is not None:
            result_df['createdAt'] = created_at
        
        return result_df
    except Exception as e:
        print("파싱 에러:", str(e))
        return None

# 첫 번째 행 파싱 시도
print("첫 번째 행 파싱 결과:")
first_row_data = parse_json(df['data'].iloc[0])

if first_row_data is not None:
    print("\n데이터 형태:", first_row_data.shape)
    print("\n처음 몇 개의 행:")
    print(first_row_data[['countryNameKr', 'currencyCode', 'basePrice']].head())
```

```python
# 전체 데이터 처리
print(f"데이터 수: {len(df)}")

# 2. 모든 행의 JSON 데이터 파싱
parsed_data = []
for idx, row in df.iterrows():
    result = parse_json(row['data'], row['createdAt'])
    if result is not None:
        # createdAt 칼럼 한국 시간대로 변경
        result['createdAt'] = pd.to_datetime(result['createdAt'], format='%Y-%m-%d %H:%M:%S') + pd.Timedelta(hours=9)
        parsed_data.append(result)

# 3. 모든 파싱된 데이터를 하나의 데이터프레임으로 합치기
final_df = pd.concat(parsed_data, ignore_index=True)

# 4. 결과 확인
print("\n최종 데이터프레임 형태:", final_df.shape)
print("\n주요 통화 환율 추이:")
main_currencies = ['USD', 'JPY', 'EUR', 'CNY']
print(final_df[final_df['currencyCode'].isin(main_currencies)][['countryNameKr', 'currencyCode', 'basePrice', 'createdAt']].head(12))
```

1. isBuyOrder = 1 일 때, price -1 이 등장하는가?
2. isBuyOrder = 0 일 때, price +1 이 등장하는가?
---

```python
# 매수 주문 필터링
buy = trade[trade['isBuyOrder'] == 1]

# 매수 주문에 대해 등장 여부 확인
matches = 0

for _, order in buy.iterrows():
    # 동일한 통화 / price +1
    match_count = filtered_df[
        (filtered_df['currencyCode'] == order['currencyCode']) &
        (filtered_df['basePrice'] == order['price'] + 1)
    ].shape[0]
    matches +=match_count

print(f'매수 주문 등장 횟수 : {matches}')
```

```python
# 더 자세한 분석을 위해 통화별로 확인
print("\n통화별 매칭 횟수:")
for currency in filtered_df['currencyCode'].unique():
    currency_buy_orders = buy[buy['currencyCode'] == currency]
    currency_matches = 0
    
    for _, order in currency_buy_orders.iterrows():
        match_count = filtered_df[
            (filtered_df['currencyCode'] == currency) & 
            (filtered_df['basePrice'] == order['price'] + 1)
        ].shape[0]
        currency_matches += match_count
    
    print(f"{currency}: {currency_matches}회")
```

```python
# 결과를 저장할 리스트 초기화
results = []

# trade의 각 행을 순회
for idx, trade_row in trade.iterrows():
    # KRW인 경우 currencyCode0 사용
    currency = trade_row['currencyCode0'] if trade_row['currencyCode'] == 'KRW' else trade_row['currencyCode']
    trade_row['executedAt'] = pd.to_datetime(trade_row['executedAt'], format='%Y-%m-%d %H:%M:%S') + pd.Timedelta(hours=9)
    # 매수 /매도에 다라 target_price 계산
    if trade_row['isBuyOrder'] == 1: # 매수
        target_price = trade_row['price'] - 1
    else: # 매도
        target_price = trade_row['price'] + 1
        # target price가 filtered_df에 존재하는지 확인
        matches = filtered_df[
            (filtered_df['currencyCode'] == currency) & 
            (filtered_df['basePrice'] == target_price)
        ].shape[0]

    
    results.append({
        'currency': currency,
        'order_type' : '매수' if trade_row['isBuyOrder'] == 1 else '매도',
        'original_price': trade_row['price'],
        'target_price': target_price,
        'found': matches > 0,
        'match_count': matches,
        'createdAt' : trade_row['executedAt']
    })

# 결과를 데이터프레임으로 변환
results_df = pd.DataFrame(results)

# 결과 요약
print("전체 거래 수:", len(results))
print("목표가가 등장한 거래 수:", results_df['found'].sum())

print("\n주문 유형별 매칭 결과:")
print(results_df.groupby('order_type')[['found']].sum())

print("\n통화별 매칭 결과:")
print(results_df.groupby('currency')[['found']].sum())

print("\n통화 및 주문 유형별 매칭 결과:")
print(results_df.groupby(['currency', 'order_type'])[['found']].sum())

# 전체 결과 확인
print("\n처음 10개 거래 상세 결과:")
results_df.head(10)
```
