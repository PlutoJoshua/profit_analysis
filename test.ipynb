```python
def fix_and_parse_json(json_string):
    try:
        # 1. 시작 부분 수정: {result" -> {"result"
        if json_string.startswith('{result"'):
            json_string = '{"result"' + json_string[8:]
        
        # 2. JSON 파싱
        data = json.loads(json_string)
        return data
    except Exception as e:
        print(f"Error: {str(e)}")
        return None

# 첫 번째 행에 대해 테스트
result = fix_and_parse_json(mama['data'].iloc[0])

if result:
    print("Successfully parsed!")
    print("\nFirst result item:")
    if 'result' in result and result['result']:
        print(json.dumps(result['result'][0], indent=2))
else:
    print("Failed to parse JSON")
```

```python
# JSON 문자열 수정 및 파싱
def parse_json(json_str):
    try:
        # 1. JSON 형식 수정
        json_str = json_str.replace('{result":', '{"result":')
        
        # 2. 마지막 따옴표 처리
        if json_str.endswith('"}'):
            json_str = json_str[:-2] + '}'
            
        # 3. JSON 파싱
        data = json.loads(json_str)
        return pd.DataFrame(data['result'])
    except Exception as e:
        print("파싱 에러:", str(e))
        # 에러 발생 위치의 문자열 출력
        if hasattr(e, 'pos'):
            pos = e.pos
            print("문제 발생 부분:", json_str[max(0, pos-50):min(len(json_str), pos+50)])
        return None

# 첫 번째 행 파싱 시도
print("첫 번째 행 파싱 결과:")
first_row_data = parse_json(df['data'].iloc[0])

if first_row_data is not None:
    print("\n데이터 형태:", first_row_data.shape)
    print("\n처음 몇 개의 행:")
    print(first_row_data.head())
```

```python
def parse_json(json_str):
    try:
        # 1. JSON 형식 수정
        json_str = json_str.replace('{result":', '{"result":')
        
        # 2. 마지막 부분 수정 (마지막 따옴표 제거)
        if json_str.endswith('}]}"'):
            json_str = json_str[:-1]  # 마지막 따옴표만 제거
            
        # 3. JSON 파싱
        data = json.loads(json_str)
        return pd.DataFrame(data['result'])
    except Exception as e:
        print("파싱 에러:", str(e))
        return None

# 첫 번째 행 파싱 시도
print("첫 번째 행 파싱 결과:")
first_row_data = parse_json(df['data'].iloc[0])

if first_row_data is not None:
    print("\n데이터 형태:", first_row_data.shape)
    print("\n처음 몇 개의 행:")
    print(first_row_data[['countryNameKr', 'currencyCode', 'basePrice']].head())
```

```python
# 전체 데이터 처리
print(f"데이터 수: {len(df)}")

# 2. 모든 행의 JSON 데이터 파싱
parsed_data = []
for idx, row in df.iterrows():
    result = parse_json(row['data'])
    if result is not None:
        result['timestamp'] = idx  # 시간 순서를 유지하기 위한 인덱스 추가
        parsed_data.append(result)

# 3. 모든 파싱된 데이터를 하나의 데이터프레임으로 합치기
final_df = pd.concat(parsed_data, ignore_index=True)

# 4. 결과 확인
print("\n최종 데이터프레임 형태:", final_df.shape)
print("\n주요 통화 환율 추이:")
main_currencies = ['USD', 'JPY', 'EUR', 'CNY']
print(final_df[final_df['currencyCode'].isin(main_currencies)][['countryNameKr', 'currencyCode', 'basePrice', 'timestamp']].head(12))
```
